```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(r'C:\Users\saeem\Desktop\Introvert\personality_synthetic_dataset.csv')

# Overview
print(df.head())
print(df.info())
print(df['personality_type'].value_counts())

# Check for missing values
print(df.isnull().sum())

# Visualize class distribution
sns.countplot(data=df, x='personality_type')
plt.title('Personality Class Distribution')
plt.show()
```
